# 2. Creating data from scratch
Pandas work with two data structure:
* Series
* DataFrame

## a. Series:
Series is a one-dimensional array object which you can say as a single column. The label asscoiated with column is called index.

```python
# it shows index on left and values on right.
# index start from 0 to N-1 (where N is the length of the data)
obj = pd.Series([-2,9,7,1])
obj
```

```python
# to get array representaion of values and index
obj.values
```

```python
obj.index
```

```python
# if you don't want default index, you can add your own label to index
obj1 = pd.Series([-2,9,7,1], index=['e','a','b','d'])
obj1
```

You can even create a series through dictionary:

```python
data = {'Burlington': 3000, 'Walmart': 4500,'Costco': 10000,'Stop & Shop':2000}
obj2 = pd.Series(data)
obj2
```

```python
data = {'Burlington': 'abcd', 'Walmart': 4500,'Costco': 10000,'Stop & Shop':2000}
obj2 = pd.Series(data)
obj2
```

## b. DataFrame:
DataFrame is a nothing but a collection of Series or a collection of columns. It is a two-dimensional structure which consist of rows and columns.

```python
data = {'stores': ['Burlington','Walmart', 'Costco', 'Stop & Shop'] , 
        'states': ['IN', 'MA','CA','NJ'],
        'sales': [100000, 40000, 70000,50000]}
frame = pd.DataFrame(data)
frame
```

you can also give your own indexes if you want instead of default, you can simply do that by adding your index name:

```python
data = {'stores': ['Burlington','Walmart', 'Costco', 'Stop & Shop'] , 
        'states': ['IN', 'MA','CA','NJ'],
        'sales': [100000, 40000, 70000,50000]}
frame = pd.DataFrame(data, index=['one','two','three','four'])
frame
```

# 3. Reading data files
We always don't create data manually with Series and DataFrame. We use already existing ddata. Data can be stored in any format: CSV, HTML, JSON, excel, SQL, etc. Below are some of the ways you can read the data:

Lets read our data. I have downloaded Walmart Sales Forecasting dataset from kaggle. I have downloaded two tables: stores.csv and train.csv
So, let's call that two dataset

```python
df1 = pd.read_csv('stores.csv')
df1.head()
```

```python
df2 = pd.read_csv('train.csv')
df2.head()
```

```python
df1.shape
```

```python
df2.shape
```

As you can see our stores dataset : df1 is of shape 45 rows and 3 columns and our train dataset: df2 is of shape 421570 rows and 5 columns. But you can see that both dataset has Store column in common. So, we can merge two dataset into one with a common column of Store.

* merge(): It is used to combine DataFarmes side by side based on common column.

```python
final_df = df1.merge(df2, on="Store")
```

```python
final_df.tail()
```

```python
final_df.shape
```

Now, we have merge two dataset and we have 421570 rows and 7 columns. In above picture I have introduces tail function
#### tail(): It displays the last 5 rows of the dataset.

```python
final_df.info()
```

```python
final_df.describe()
```

```python
final_df.columns
```

# 7. Adding a column in a data frame and using of mapping method

```python
# adding new row month and year
final_df['month'] = pd.DatetimeIndex(final_df['date']).month
final_df['year'] = pd.DatetimeIndex(final_df['date']).year
final_df.tail()
```

```python
final_df
```

```python
#for mapping method creating dictionary
month_nummber_to_name = {
      1: 'January',
      2: 'February',
      3: 'March',
      4: 'April',
      5: 'May',
      6: 'June',
      7: 'July',
      8: 'August',
      9: 'September',
      10: 'October',
      11: 'November',
      12: 'December'
}
```

```python
#created new column month_name using mapping method
final_df['month_name'] = final_df['month'].map(month_nummber_to_name)
```

```python
final_df.tail()
```
